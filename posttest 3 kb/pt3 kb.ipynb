```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np

ibnu = pd.read_csv("bank-full.csv")
ibnu.head()
```

pertama kita import library yang mau kita pakai dan kita tampilkan csv nya

```python
ibnu.set_index('job', inplace=True)
ibnu.head()
```

```python
status_pekerja = ibnu['marital'].value_counts()
status_pekerja
```

kita membuat variabel status pekerja agar mudah dipanggil

```python
status_pekerja.plot.bar()
plt.title('Data Total dan Status Pekerja')
plt.xlabel('Status Pekerja')
plt.ylabel('Total Pekerja')
plt.show()
```

disini kita menampilkan grafik batang tentang status para pekerja yang sudah menikah atau belum ,yang sudah menikah 25000+

```python
plt.hist(ibnu['marital'])
plt.title('Data Total dan Status Pekerja')
plt.xlabel('Status Pekerja')
plt.ylabel('Total Pekerja')
plt.show()
```

disini kita memakai histogram untul mengkalkulasi status para pekerja

```python
Pekerja_group = ibnu.groupby('marital').sum()
Pekerja_group.head()
```

disini kita membuat variabel baru agar mudah dipanggil

```python
plt.figure(figsize=(7,12))
Pekerja_group['age'].plot.pie(autopct='%.1f%%')
plt.title('Total pekerja jika dikonversi ke persenan')
```

disini kita membuat pie chart membuat persenan status para pekerja

```python
plt.figure(figsize=(15, 6)) 
sns.scatterplot(
    data=ibnu,
    x='job',
    y='duration',
)

plt.title('Data durasi perkerja menurut pekrjannya')
plt.xlabel('pekerjaan')
plt.ylabel('durasi')
plt.show()
```

disini kita membuat grafik scatterplot terlihat smua pekerja beserta durasinya

```python
ibnu = np.random.rand(5, 5)
sns.heatmap(ibnu, annot=True, cmap='coolwarm')
plt.xlabel('duration')
plt.ylabel('job')
plt.title('pekerjaan beserta durasinya')
plt.show()
```
